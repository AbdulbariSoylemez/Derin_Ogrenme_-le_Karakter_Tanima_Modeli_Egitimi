```python
# Fotograf verilerimizi adresini ve etiketini bir csv dosyaysına çeviriyoruz 

import os
import csv

# Fotoğraf dosyalarının bulunduğu dizin yolu
#foto_dizin_yolu = "/Users/abdulbarisoylemez/Documents/Visual Code/Bitirme Projesi/kod/data/isaretData"

# CSV dosyasının adı ve yolu
csv_dosya_yolu = "fotograf_etiketler.csv"

# Eğer etiketler metin şeklindeyse ve "A" ise 0, "B" ise 1, ... gibi bir haritalama yapın
label_mapping = {"A": 0, "B": 1, "Bosluk": 2, "C": 3, "D": 4, "DOT": 5, "E": 6, "F": 7, "G": 8, "H": 9, "I": 10, "K": 11, "L": 12, "M": 13, "N": 14, "O": 15, "P": 16, "R": 17, "S": 18, "SPACE": 19, "T": 20, "U": 21, "V": 22, "Y": 23, "Z": 24}

# CSV dosyasını yazma modunda aç
with open(csv_dosya_yolu, 'w', newline='') as csvfile:
    csv_writer = csv.writer(csvfile)
    
    # CSV dosyasının başlıklarını yaz
    csv_writer.writerow(["Dosya Yolu", "Etiket"])
    
    # Dizin içindeki her dosyayı dolaş
    for dosya in os.listdir(foto_dizin_yolu):
        dosya_yolu = os.path.join(foto_dizin_yolu, dosya)
        
        # Dosya adından etiketi ayıkla ve haritalamadaki karşılığını al
        etiket = dosya.split("_")[0]
        sayisal_etiket = label_mapping.get(etiket, -1)  # Eğer etiket haritalamada yoksa -1 döndür
        
        # Dosya yolu ve sayısal etiketi CSV dosyasına yaz
        csv_writer.writerow([dosya_yolu, sayisal_etiket])

print(f"{len(os.listdir(foto_dizin_yolu))} dosya yolunu ve sayısal etiketi '{csv_dosya_yolu}' CSV dosyasına kaydettim.")
```
